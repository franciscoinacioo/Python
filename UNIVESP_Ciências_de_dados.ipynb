### Diferente de uma lista python, um recorte (fatia) de um ndarray sofrerá broadcast - é como se fosse uma view de bando de dados.
### Com conjuntos imensos de dados isso é uma grande vantagem em relação a performance

```python
dadosseq[3:6] #posição 6 não entra
```

```python
dadosseq[3:6] = 20
```

```python
dadosseq
```

```python
fatiadados=dadosseq[2:4]
```

```python
fatiadados
```

```python
fatiadados[1] = 2000
```

```python
fatiadados
```

```python
dadosseq #Como fiz copia (fatiadados=dadosseq[2:4]) ele vai remeter no valor original  ------> broadcast
```

```python
fatiadados[:] = 333 # [:] -----> pego todos os valores no caso de fatiadados do 2 até o 4, dois valores
```

```python
dadosseq
```

# Transposição
