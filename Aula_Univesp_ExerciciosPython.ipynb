Funções

```python
def exercicio1():
  palavra = input(str("Digite uma palavra: "))
  lista = list(palavra)
  segredo = ""
  for x in lista:
    if (x=='z'): segredo+='a'       #segredo - variável auxiliar
    elif (x=='Z'): segredo+='A'
    else: segredo+=chr(ord(x)+1)     #substitui pela proxima letra do alfabeto
  print(segredo)
```

```python
exercicio1()
```

```python
import pandas as pd
uri = "https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv"
bebidas = pd.read_csv(uri)
bebidas.head()
```

```python
bebidas.columns = ["país", "cervejas", "destilados", "vinhos", "total"]
bebidas.head()
```

```python
bebidas['cervejas'].mean()
```

```python
bebidas['cervejas'].std()
```

```python
bebidas['cervejas'].median()
```
